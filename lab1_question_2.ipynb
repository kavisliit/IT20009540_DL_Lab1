```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Read the CSV file into a DataFrame
input_file_path = '/content/drive/MyDrive/Au_nanoparticle_dataset.csv'
df = pd.read_csv(input_file_path)
```

```python
columns_to_keep = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

new_df = df[columns_to_keep]

print(new_df.head(20))
```

```python
mean_values = new_df.mean()
std_values = new_df.std()
quartile_values = new_df.describe(percentiles=[.25, .50, .75])

# Display the calculated statistics
print("Mean values:\n", mean_values)
print("\nStandard deviation values:\n", std_values)
print("\nQuartile values:\n", quartile_values)
```

```python
# Plot histograms of each feature in a 1x4 layout
plt.figure(figsize=(15, 5))

for i, column in enumerate(columns_to_keep):
    plt.subplot(1, 4, i+1)
    new_df[column].hist(bins=20, color='blue', alpha=0.7)
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create pair plots using seaborn
sns.set(style="ticks")
sns.pairplot(new_df)
plt.show()
```

```python
#new_df is the dataframe containing only the above mentioned 4 features.
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
```
