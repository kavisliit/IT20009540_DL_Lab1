```python
import numpy as np
import matplotlib.pyplot as plt

# Create a random 100000x1 array from exponential distribution
scale_parameter = 2.0
random_array_exp = np.random.exponential(scale=scale_parameter, size=(100000, 1))

# Create a random 100000x1 array from uniform distribution
random_array_uniform = np.random.rand(100000, 1)

# Create a random 100000x1 array from normal distribution
random_array_normal = np.random.normal(size=(100000, 1))

# Plot histograms
plt.hist(random_array_exp, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(random_array_uniform, density=True, bins=50, histtype="step", color="orange", label="Uniform")
plt.hist(random_array_normal, density=True, bins=50, histtype="step", color="green", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate X and Y values within the specified range
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z values using the function Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot of Z = X**2 + Y**2')

# Show the plot
plt.show()
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Display first 5 observations
df.head()
```

```python
# Calculate Pearson and Spearman correlations
correlation_pearson = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')
correlation_spearman = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

# Set up the figure
plt.figure(figsize=(12, 6))

# Plot the Pearson correlation heatmap
plt.subplot(1, 2, 1)
sns.heatmap(correlation_pearson, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Pearson Correlation')

# Plot the Spearman correlation heatmap
plt.subplot(1, 2, 2)
sns.heatmap(correlation_spearman, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Spearman Correlation')

# Adjust layout
plt.tight_layout()

# Show the heatmaps
plt.show()
```
